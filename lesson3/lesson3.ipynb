```python
import pandas as pd 
import numpy as np 
```

```python
students_performance = pd.read_csv('StudentsPerformance.csv')
students_performance = students_performance.rename(columns={
'parental level of education':'parental_level_of_education',
'test preparation course':'test_preparation_course',
'math score':'math_score',
'reading score':'reading_score',
'writing score':'writing_score'
})
```

```python
students_performance.groupby(['gender', 'race/ethnicity'], as_index=False).aggregate({'math_score':'mean', 'reading_score':'mean'}).rename(columns = {
    'math_score' : 'mean_math_score',
    'reading_score' : 'mean_reading_score'
})
```

```python
mean_scores = students_performance.groupby(['gender', 'race/ethnicity']).aggregate({'math_score':'mean', 'reading_score':'mean'}).rename(columns = {
    'math_score' : 'mean_math_score',
    'reading_score' : 'mean_reading_score'
})
```
